#### Création des variable :
- 'GFAPerBuilding' : calcule la surface de chaque batiment.
- 'GFAPerFloor' : calcule la surface de chaque étage.
- 'GFA_buildingrate' : pourcentage de la surface du batiment par rapport à la surface totale du batiment et du parking 

```python
df.drop(['LargestPropertyUseTypeGFA', 
           'SecondLargestPropertyUseTypeGFA',
           'SecondLargestPropertyUseType',
           'ThirdLargestPropertyUseTypeGFA',
           'ThirdLargestPropertyUseType',
           'PropertyGFAParking'
           ],
         axis=1, inplace=True)
df['GFAPerBuilding'] = round((df['PropertyGFATotal'] / df['NumberofBuildings']),3)
df['GFAPerFloor'] = round((df['PropertyGFATotal'] / df['NumberofFloors']),3)

```

```python
df['GFA_buildingrate'] = round(df['PropertyGFABuilding(s)'].fillna(0)/df['PropertyGFATotal'].fillna(0), 5)
```
